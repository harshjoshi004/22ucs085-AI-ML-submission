```python
cdf=df.copy()
cdf.isnull().sum()
```

```python
m = cdf["R&D Spend"].mean()
s = cdf["R&D Spend"].std()
z = [ ((i-m)/s) for i in cdf["R&D Spend"]]
outliner= [ i for i in z if (i > 3) or ( i<-3)]
if(outliner==[]):
  print("There is no outliner in R&D Spend")
```
